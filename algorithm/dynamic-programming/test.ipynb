```python
from math import floor

def numSquares( n: int) -> int:
        clps = int(floor(n**(0.5)))
        egele = []
        for i in range(1,clps+1):
            egele.append(i**2)
        def solve(k,eg,memo={}):  
            if(k in memo):
                return memo[k]
            if(k < 0):
                return 0
            if(k == 0):
                return 1
            minNo = []
            
            for i in eg:
                mm = solve(k-i,eg,memo)

                if(mm):
                    minNo.append(mm)
                    

            if(minNo):
                memo[k] = min(minNo)+1
            else:
                memo[k] = 0
            return memo[k]
        return solve(n,egele) 
print(numSquares(13))
```

```python
from collections import Counter

def diff(s1,s2):
    n = len(s1)
    i = 0
    d = 0
    while(i<n):
        if(s1[i] != s2[i]):
            d+=1
        i+=1
    return d


def segAB(A):

    conts = Counter(A)

    s1 = "a"*conts["a"] + "b"*conts["b"]
    s2 = "b"*conts["b"] + "a"*conts["a"]

    ans = min(diff(s1,A),diff(s2,A))
    return ans


n = int(input())
s = input()
print(segAB(s))
```

```python

def bfsUtil(v, vis, adj, src):
	vis[src] = True
	q = []
	q.append(src)
	
	count = 0
	while (len(q) != 0):
		p = q[0]
		
		for i in range(len(adj[p])):
			if (vis[adj[p][i]] == False and v[adj[p][i]] == 0):
				count += 1
				vis[adj[p][i]] = True
				q.push(adj[p][i])
			
		
			elif(vis[adj[p][i]] == False and v[adj[p][i]] == 1):
				count += 1
		q.remove(q[0])
	
	return count + 1


def minimumCost(n,m, cost, connections):
	vis = [ 0 for i in range(n + 1)]
	mx = 1

	for i in range(1, n + 1, 1):
		vis[i] = False
		

	for i in range(1, n + 1, 1):
		
		if (cost[i] == 0 and vis[i] == False):
			res = bfsUtil(cost, vis, connections, i)
			if (res > mx):
				mx = res

	return mx
```
